```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sigmoid Activation Function
# Formula: sigmoid(x)= 1 / (1 + exp(-x))

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

inputs= np.array([-1, -0.5, 0, 0.5, 1])
print(f"Sigmoid: {sigmoid(inputs)}")
```

```python
plt.plot(inputs, sigmoid(inputs))
plt.title('Sigmoid')
plt.show()
```

```python
# Tanh Activation Function
# Formula: tanh(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x))

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

inputs= np.array([7, 12, -11, -3, 33])
print(f"Tanh: {tanh(inputs)}")
```

```python
plt.plot(inputs, tanh(inputs))
plt.title('Tanh')
plt.show()
```

```python
# ReLU Activation Function
# Formula: relu(x) = max(0,x) 

def relu(x):
    return np.maximum(0,x)

inputs= np.array([44, -2, -23, 11, 17])
print(f"Relu: {relu(inputs)}")
```

```python
plt.plot(inputs, relu(inputs))
plt.title('Relu')
plt.show()
```

```python
# LeakyRelu Activation Function
# Formula: leaky_relu(x) = x if x > 0 else alpha*x (alpha=0.01)

def leaky_relu(x, alpha = 0.01):
    return np.where(x>0, x, alpha*x)

inputs= np.array([-32, 31, 17, -12, 3])
print(f"Leaky Relu: {leaky_relu(inputs)}")
```

```python
plt.plot(inputs, leaky_relu(inputs))
plt.title('Leaky Relu')
plt.show()
```

```python
# Softmax Activation Function
# Formula: softmax(x) = exp(x) / sum(exp(x)) 

def softmax(x):
    e_x= np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

inputs= np.array([1, -19, 37, -28, -16])
print(f"Softmax: {softmax(inputs)}")
```

```python
plt.plot(inputs, softmax(inputs))
plt.title('Softmax')
plt.show()
```

```python
activations={
    "Sigmoid" : sigmoid,
    "Tanh" : tanh,
    "Relu" : relu,
    "Leaky_Relu" : leaky_relu,
    "Softmax" : softmax
}
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,20))
for i, (name,func) in enumerate(activations.items(),1):
  plt.subplot(5,2,i)
  plt.plot(inputs,func(inputs))
  plt.title(name)
  
plt.tight_layout()
plt.show
```
